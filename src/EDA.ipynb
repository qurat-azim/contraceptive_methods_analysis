```python
import pandas as pd
import numpy as np
import altair as alt
import altair_ally as aly
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/raw/cmc_data.txt',names=['Age','Education','Spouse Education','Children',
                                                'Religion','Work','Spouse Occupation','Living Standard',
                                                'Media Exposure','Contraceptive Method'])
df.head()
```

```python
df['Contraceptive Method'] = df['Contraceptive Method'].apply(str)
df['Spouse Occupation'] = df['Spouse Occupation'].apply(str)
```

```python
df['Contraceptive Method'].value_counts()
```

```python
alt.Chart(df).mark_bar().encode(
    x='Work:N',
    y='mean(Age):Q',
    color='Contraceptive Method:N',
    column='Contraceptive Method:N'
).properties(width=50)
```

```python
alt.Chart(df).mark_boxplot(extent='min-max').encode(
    y='Contraceptive Method:O',
    x='Age:Q',
    color='Contraceptive Method:N'
).properties(height=80)
```

```python
alt.Chart(df).mark_boxplot(extent='min-max').encode(
    y='Contraceptive Method:O',
    x='Children:Q',
    color='Contraceptive Method:N'
).properties(height=80)
```

```python
alt.Chart(df).mark_bar().encode(
    x='Living Standard:N',
    y='count()',
    color='Contraceptive Method:N'
    
).properties(width=80)
```
